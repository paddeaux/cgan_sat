```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

gen_loss = pd.read_csv('colour_overfit_losses/simple_gen_loss_history.csv')

critic_loss = pd.read_csv('colour_overfit_losses/simple_discrim_loss_history.csv')

fig, ax = plt.subplots(1,1,figsize=(10,8))

gen_loss.plot.scatter(ax=ax, x='batch', y='gen_loss', c='orange',s=0.4)
critic_loss.plot.scatter(ax=ax, x='batch', y='discrim_loss', c='blue',s=0.4)
plt.show()

```
