```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

gen_loss = pd.read_csv('/home/paddy/git/losses/cgan_gen_loss_epoch2.csv')

critic_loss = pd.read_csv('/home/paddy/git/losses/cgan_critic_loss_epoch2.csv')

fig, ax = plt.subplots(1,1,figsize=(10,8))

gen_loss.plot.scatter(ax=ax, x='batch', y='gen_loss', c='blue',label="Generator", s=0.6)
critic_loss.plot.scatter(ax=ax, x='batch', y='discrim_loss', c='orange', label="Critic", s=0.6)
plt.xlabel("Batch Number")
plt.ylabel("Loss")

ax.legend()

plt.show()

```
